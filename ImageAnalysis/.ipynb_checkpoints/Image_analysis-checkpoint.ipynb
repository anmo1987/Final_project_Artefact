```python
###Prepare dataset for pca
pca_data = data_filt
var_to_keep = ["dx", "dx_type", "age", "sex", "localization"]
pca_data = pca_data[var_to_keep]
pca_data
```

```python
pca_data["localization"].value_counts()
```

```python
##Replace qualiative variable to quanti
lesion_type_dict = {
    'nv': '1',
    'mel': '2',
    'bkl': '3',
    'bcc': '4',
    'akiec': '5',
    'vasc': '6',
    'df': '7'
}
pca_data['disease'] = pca_data['dx'].map(lesion_type_dict.get) 
pca_data["sex"] = pca_data["sex"].map({'female':0, 'male':1})
pca_data["dx_type"] = pca_data["dx_type"].map({'histo':0, 'follow_up':1, 'consensus':2, 'confocal':3})
pca_data["localization"] = pca_data["localization"].map({'back':0, 'lower extremity':1, 'trunk':2, 'upper extremity':3,
                                                        'abdomen':4,'face':5,'chest':6,
                                                        'foot':7,'unknown':8,'neck':9,'scalp':10,'hand':11,'ear':12,'genital':13,'acral':14})
pca_data_1 = pca_data.drop("dx", axis=1)
```

```python
pca_data_1 = pca_data.drop("dx", axis=1)
y = pca_data["dx"]
```

```python
##Compute PCA
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

pipeline = Pipeline([("scaler", StandardScaler()), ("pca", PCA(n_components=2)),])
res_pca_data = pd.DataFrame(pipeline.fit_transform(pca_data_1), columns=["PC1", "PC2"])
#pca_data["species"] = iris_df["species"]

pca_step = pipeline.steps[1][1]
loadings = pd.DataFrame(
    pca_step.components_.T,
    columns=["PC1", "PC2"],
    index=pca_data_1.columns,
)
```

```python
def loading_plot(
    coeff, labels, scale=1, colors=None, visible=None, ax=plt, arrow_size=0.5
):
    for i, label in enumerate(labels):
        if visible is None or visible[i]:
            ax.arrow(
                0,
                0,
                coeff[i, 0] * scale,
                coeff[i, 1] * scale,
                head_width=arrow_size * scale,
                head_length=arrow_size * scale,
                color="#000" if colors is None else colors[i],
            )
            ax.text(
                coeff[i, 0] * 1.15 * scale,
                coeff[i, 1] * 1.15 * scale,
                label,
                color="#000" if colors is None else colors[i],
                ha="center",
                va="center",
            )


g = sns.scatterplot(data=res_pca_data, x="PC1", y="PC2", hue=y)

# Add loadings
loading_plot(loadings[["PC1", "PC2"]].values, loadings.index, scale=8, arrow_size=0.02)


# Add variance explained by the
g.set_xlabel(f"PC1 ({pca_step.explained_variance_ratio_[0]*100:.2f} %)")
g.set_ylabel(f"PC2 ({pca_step.explained_variance_ratio_[1]*100:.2f} %)")

plt.savefig("PCA_with_loadings.png", dpi=200)
plt.show()
```

TESTING CLASSICAL MACHINE LEARNING FOR CATEGORIZED DATA FOR DISEASE PREDICTION

```python
##PREPARING DATASET
ml_data = data_filt
var_to_keep = ["dx", "dx_type", "age", "sex", "localization"]
ml_data = ml_data[var_to_keep]
ml_data['disease'] = ml_data['dx'].map(lesion_type_dict.get) 
ml_data["sex"] = ml_data["sex"].map({'female':0, 'male':1})
ml_data
```

```python
###Using get dummies for categorical dataset
X_dx_type = pd.get_dummies(ml_data['dx_type'], dtype=float)
X_localization = pd.get_dummies(ml_data['localization'], dtype=float)
ml_data = ml_data.drop(["dx", 'dx_type', 'localization'], axis=1)
```

```python
##Concat dataset
ml_final = pd.concat([X_dx_type, X_localization, ml_data], axis=1)
ml_final['disease'] = ml_final['disease'].astype(int)
```

```python
##Save ML final dataset
ml_final.to_csv('ML_dataset.csv')
```

```python
###prepare X & y train & split data
y = ml_final["disease"]
X= ml_final.drop("disease", axis=1)
```
