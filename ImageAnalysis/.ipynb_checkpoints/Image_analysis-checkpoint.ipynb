```python
##Missing data handling
##Filling with Mean by sex
data['age'] = data['age'].fillna(data.groupby(['sex'])['age'].transform('mean'))
data.isna().sum()
```

```python
##Create a synthetized data frame
df = data.groupby(['sex', 'dx', 'localization']).size().unstack(fill_value=0)
mean_by_group = pd.DataFrame(data.groupby(['sex', 'dx'])['age'].mean())
age = mean_by_group["age"]
synthetized = pd.concat([df, age], axis=1)
synthetized["sum"] = synthetized["abdomen"] + synthetized["acral"] + synthetized["back"] +synthetized["chest"]+synthetized["ear"]+synthetized["face"]+ synthetized["foot"]+ synthetized["genital"]+ synthetized["hand"]+ synthetized["lower extremity"]+synthetized["neck"]+ synthetized["scalp"]+ synthetized["trunk"]+ synthetized["unknown"]+ synthetized["upper extremity"]
```

```python
###Prepare dataset for pca
pca_data = data_filt
var_to_keep = ["dx", "dx_type", "age", "sex", "localization"]
pca_data = pca_data[var_to_keep]
pca_data
```

```python
pca_data["localization"].value_counts()
```

```python
##Replace qualiative variable to quanti
lesion_type_dict = {
    'nv': '1',
    'mel': '2',
    'bkl': '3',
    'bcc': '4',
    'akiec': '5',
    'vasc': '6',
    'df': '7'
}
pca_data['disease'] = pca_data['dx'].map(lesion_type_dict.get) 
pca_data["sex"] = pca_data["sex"].map({'female':0, 'male':1})
pca_data["dx_type"] = pca_data["dx_type"].map({'histo':0, 'follow_up':1, 'consensus':2, 'confocal':3})
pca_data["localization"] = pca_data["localization"].map({'back':0, 'lower extremity':1, 'trunk':2, 'upper extremity':3,
                                                        'abdomen':4,'face':5,'chest':6,
                                                        'foot':7,'unknown':8,'neck':9,'scalp':10,'hand':11,'ear':12,'genital':13,'acral':14})
pca_data_1 = pca_data.drop("dx", axis=1)
```

```python
##Compute PCA
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

pipeline = Pipeline([("scaler", StandardScaler()), ("pca", PCA(n_components=4)),])
pca_data = pca_data.drop("dx", axis=1)
pca_data = pd.DataFrame(pipeline.fit_transform(pca_data.drop("dx", axis=1)), columns=["PC1", "PC2", "PC3", "PC4"])
#pca_data["species"] = iris_df["species"]

pca_step = pipeline.steps[1][1]
loadings = pd.DataFrame(
    pca_step.components_.T,
    columns=["PC1", "PC2"],
    index=data.columns,
)
```
